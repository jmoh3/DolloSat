# Let's take a look at k-Dollo Phylogenies!

For this walkthrough we'll take a look at all the examples in the data/small data directory.

```python
import os

directory_name = 'data/small_data'

directory = os.fsencode('data/small_data')
files = []

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".txt"):
        files.append(f'data/small_data/{filename}')
    else:
        continue
```

Now, we'll generate the cnf formulas for each of the input matrices and record how long it takes to generate formulas for each of the inputs. Just a warning, this will take a while.

```python
from solve_k_dollo.generate_formula import read_matrix, generate_cnf
import time

def get_info(filename):
    split_filename = filename.split('/')[2].split('_')
    short_filename = filename.split('/')[2].split('.')[0]
    
    rows = split_filename[0][1:]
    columns = split_filename[1][1:]
    
    if int(rows) < 100 and int(columns) < 100:
        start = time.time()
        matrix = read_matrix(filename)
        num_clauses = generate_cnf(matrix, f'formulas/{short_filename}.formula.cnf')
        end = time.time()

        elapsed = end - start

        return f'{short_filename}.txt,{rows},{columns},{elapsed},{num_clauses}'
    else:
        return None
```

```python
all_info = []

for file in files:
    info = get_info(file)
    if info:
        all_info.append(info)
```

Now, let's save this data in a csv, then load it into a dataframe before we add to it more later.

```python
for i in range(len(all_info)):
    all_info[i] += '\n'
```

```python
out = 'filename,rows,columns,time,clauses\n'

for info in all_info:
    out += info
```

```python
with open('formula_generation_data.csv', 'w') as f:
    f.write(out)
```

```python
import pandas as pd

df = pd.read_csv('formula_generation_data.csv')
```

```python
df
```

We can plot the time it takes to create the cnf formula as a function of the number of rows in the input matrix.

This is generally the behavior we'd expect, considering it takes O(m^3) time to generate the cnf formula with respect to the number of rows (assuming # of columns is constant).

```python
df[df['columns'] == 10].plot.scatter(x='rows', y='time')
```

We can do the same for columns. Because the runtime for formula generation grows at a rate of O(n^2) with respect to the number of columns, this is what we'd expect.

```python
df[df['rows'] == 10].plot.scatter(x='columns', y='time')
```

Now, we'll take a look at generating the total number of solutions for a given input matrix. Because this is so time intensive, we'll only look at our 5x5 samples

```python
directory = os.fsencode('data/small_data')
files = []

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".txt"):
        split_filename = filename.split('_')
        rows = split_filename[0][1:]
        columns = split_filename[1][1:]
        if rows == '5' and columns == '5':
            files.append(f'data/small_data/{filename}')
    else:
        continue
```

```python
files
```

`brute_force_solver` will give us the total number of solutions for a given binary matrix. Another warning, this will take a while - a lot longer than the last step

```python
from solve_k_dollo.brute_force_solver import find_all_solutions
num_solutions = {}

for file in files:
    matrix = read_matrix(file)
    solutions = find_all_solutions(matrix, None, False)
    num_solutions[file] = solutions
    print(solutions)
```

```python
num_solutions
```

Let's select the subset of rows in our dataframe where the number of rows and columns in the input matrix is equal to 5, and add total solutions as a column for this dataframe.

```python
min_df = df[df['rows'] == 5][df['columns'] == 5]
```

```python
min_df['total_solutions'] = min_df['filename'].apply(lambda x: num_solutions['data/small_data/'+x])
```

And save it to a csv for later!

```python
min_df.to_csv('small_examples.csv')
```

Now, let's run our tool on thse input matrix to see how many solutions Quicksampler finds. We'll do this using the shell script `generate_samples_many.sh`, which will save the number of solutions found by Quicksampler to a text file.

```python
os.system('sh generate_samples_many.h {directory_name}')
```

Now, we'll read in the number of samples generated by Quicksampler and add them to our dataframe.

```python
sampled = None

with open('num_solutions.txt', 'r') as f:
    sampled = f.readlines()

sampled_dict = {}

for sample in sampled:
    num_samples = sample.split()[0]
    file = sample.split()[1].split('/')[2].split('.')[0] + '.txt'
    sampled_dict[file] = num_samples
```

```python
min_df['total_samples'] = min_df['filename'].apply(lambda x: sampled_dict[x])
```

And then we can calculate the percentage of solutions found by Quicksampler

```python
min_df['percent_found'] = min_df['total_samples'].astype(int) / min_df['total_solutions'].astype(int)
```
